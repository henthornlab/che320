## Lecture 8 - Multiple Resistances

We rarely see heat or mass transfer occur through a single phase... rather, we see multiple materials or media interacting. This sets up a situation where the rate of heat or mass transfer is affected by not just one phase, but it is the cumulative effect of all phases and interfaces.

Take for instance the steady state heat loss from a warm ($T = T_H$) room through a window out to the cold ($T = T_C$) environment. We have, in essence two systems:

1. Heat transfer in the air of the room to the inner surface of the window.
2. Heat transfer through the window to an outer temperature of $T = T_C$.


Since we focused on transfer through a solid material like a window previously, let's start there.



# Temperature profile through the window

We have solved for condution through the window previously.

Briefly, we simplified Fourier's 2nd law from:
$$\rho C_P \frac{\partial T}{\partial t} = k\frac{\partial^2 T}{\partial x^2} + \dot q $$

to:
$$0 = k\frac{\partial^2 T}{\partial x^2} $$

and:

$$\frac{d^2 T}{d x^2} = 0$$

In this case, we will solve on the domain of: $$0 \leq x \leq L $$ and apply the steady state boundary conditions of: $$T(0) = T_0$$ and $$T(L) = T_C$$

The solution to this is:

$$T(x) = (T_L-T_C) \frac{x}{L} + T_0  $$


# Heat flux through the window

Since we are working with conduction through the window:

$$q^"_{win} = -k\frac{dT}{dx}$$  

which becomes:
$$q^"_{win} = -k\frac{d}{dx}\left( (T_C-T_0) \frac{x}{L} + T_0 \right)$$  

$$q^"_{win} = -k \frac{(T_C-T_0)}{L} $$  

# Heat flux in the room

Assuming convective heat transfer in the room from the hot center ($T=T_H$) to the window's surface ($T = T_0$), we should see:

$$q^"_{in} = h_{in} (T_H - T_0)$$



# Putting it all together

At steady state, the rate of heat flowing from the room to the window should be equal. 

$$\dot q_{in} = \dot q_{win} $$


We can relate the rate of heat flow $\dot q$ to to the fluxes by carefully considering the area, $\dot q = q^" A$.

In this simple geometry, the window area is the same on the inside and the out.

$$q^"_{in}A = q^"_{win}A$$

or

$$q^"_{in} = q^"_{win}$$

# Solving for the combined heat flux

Substituing the heat fluxes with their convective and conductive forms and noting that all heat fluxes are the same:

$$q^" = h_{in} (T_H - T_0)$$
$$q^" = -k \frac{(T_C-T_0)}{L} $$  

We get:

$$h_{in} (T_H - T_0) = -k \frac{(T_C-T_0)}{L}$$

Most likely, however, we know $T_H$ and $T_C$, not $T_0$.

Let's solve for those to get rid of them.


Rearrange the heat flux in the window to solve for $T_0$ since we wish to get rid of it.


$$q^" = k \frac{(T_0 - T_C)}{L} $$ 
$$\frac{q^"L}{k} = T_0 - T_C $$ 
$$ T_0 = T_C + \frac{q^"L}{k}$$ 

$$q^" = h_{in} (T_H - T_0) = h_{in} (T_H - T_C - \frac{q^"L}{k} )$$
$$q^" +  h_{in} \frac{q^"L}{k}=  h_{in} (T_H - T_C)$$

$$q^" \left( 1 +   \frac{h_{in} L}{k} \right)=  h_{in} (T_H - T_C)$$
$$q^"=  \frac{h_{in}}{\left( 1 +   \frac{h_{in} L}{k} \right)} (T_H - T_C)$$
$$q^"=  \frac{1}{\left( \frac{1}{h_{in}} +   \frac{ L}{k} \right)} (T_H - T_C)$$



# The overall heat transfer coefficient

We've now solved for the heat transfer flux through two distinct media, and we arrive at a unique form for rectangular (x,y,z) coordinates:

$$q^"=  \frac{1}{\left( \frac{1}{h_{in}} +   \frac{ L}{k} \right)} (T_H - T_C)$$

or,

$$q^"=  U (T_H - T_C)$$

with $U$, ***the overall heat transfer coefficient*** in rectangular coordinates, defined as:

$$ U = \frac{1}{\left( \frac{1}{h_{in}} +   \frac{ L}{k} \right)}$$

The form is extremely interesting... it is the reciprocal of the sum of reciprocals of the the conductances of each phase.



# General form of the overall heat transfer coefficient

We can repeat this process for more materials/layers/etc. The result, however, is always the same:

$$ U = \left( Conductance_1^{-1} + Conductance_2^{-1} + Conductance_3^{-1} + ...\right)^{-1}$$

Or alternatively, we see that it is the ***reciprocal of the sum of the heat transfer resistances***...

$$ U = \left( Resistance_1 + Resistance_2 + Resistance_3 + ...\right)^{-1}$$

In rectangular coordinates, these resistances are either:

* Convective: $1/h$
* Conductive: $L/k$



Example:

We have four materials joined for heat transfer, with parameters:
$h_1, k_2, L_2, k_3, L_3$, and $h_4$

The overall heat transfer coefficient will therefore be:

$$ U = \frac{1}{\left( \frac{1}{h_1} +   \frac{ L_2}{k_2} +   \frac{ L_3}{k_3} + \frac{1}{h_4}\right)}$$
