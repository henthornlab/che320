```
import plotly.express as px
import pandas as pd

df1 = pd.read_csv('https://raw.githubusercontent.com/henthornlab/data/master/steam-800.csv')

fig = px.scatter_3d(df1, x='Temperature (C)', y='Pressure (psia)', z='Density (kg/m3)',
              color='Phase')
fig.show()
```

```
fig2 = px.scatter_3d(df1, x='Temperature (C)', y='Pressure (psia)', z='Enthalpy (kJ/kg)',
              color='Phase')
fig2.show()
```

```
fig3 = px.scatter_3d(df1, x='Temperature (C)', y='Pressure (psia)', z='Cv (J/g*K)',
              color='Phase')
fig3.show()
```
